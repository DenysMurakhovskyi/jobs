```python
with open('skill_level_dict_old.pkl', 'rb') as f:
    sld = pkl.load(f)
```

```python
jobs_old = pd.read_csv('jd_expanded_old.csv')
```

```python
jobs_new = pd.read_excel('job_data_all (2).xlsx')

jobs_new['id'] = range(len(jobs_new))
jobs_new['time'] = '2021-01-01/01/21'
jobs_new = jobs_new.merge(companies_final[['Company_Name', 'company_id']])
jobs_new['job_salary'] = jobs_new.Price.apply(lambda x: x.split('-')[0] + 'k' + '-' + x.split('-')[-1] + 'k')
jobs_new['job_temptation'] = str([])
jobs_new['skill_set'] = jobs_new.Skills.apply(lambda x: str(x.replace("'", '"').replace(' ', '').replace(',]', ']').lower()))
jobs_new['comp_id'] = jobs_new.company_id
jobs_new['job_title'] = jobs_new.Job_Title
jobs_new['job_address'] = jobs_new.Location
```

```python
res = list()
skill_set = set()
level_set = set()

for k in jobs_new.skill_set:
    qq = json.loads(k)
    qq_new = list()
    for t in qq:
        if t['skill'] in sld[0].keys():
            qq_new.append(t)
            skill_set.add(t['skill'])
            level_set.add(t['level'])
    res.append(str(qq_new).replace("'", '"'))
```

```python
sld_new_0 = dict()

for k, v in sld[0].items():
    if k in skill_set:
        sld_new_0[k] = v
```

```python
sld_new_1 = dict(list(zip(list(set(sld_new_0.values())), range(len(set(sld_new_0.values()))))))
```

```python
sld_new_2 = dict()

for k, v in sld[2].items():
    if k in level_set:
        sld_new_2[k] = v
```

```python
sld_new_3 = dict(list(zip(list(level_set), range(len(level_set)))))
```

```python
cities_old = pd.read_csv('../city_features_old.csv')
```

```python
jobs_new['price_low'] = jobs_new.Price.apply(lambda x: int(x.split('-')[0]))
jobs_new['price_high'] = jobs_new.Price.apply(lambda x: int(x.split('-')[1]))
jobs_new['price_avg'] = (jobs_new['price_low'] + jobs_new['price_high']) / 2
```

```python
# 50th Percentile
def q25(x):
    return x.quantile(0.25)
# 90th Percentile
def q75(x):
    return x.quantile(0.75)
```

```python
city_features = jobs_new[['Location', 'price_low', 'price_avg', 'price_high']].groupby('Location').agg([len, q25, np.median, q75, np.mean, np.std]).reset_index()
```

```python
city_features.columns = ['city', 'jd_count', 'low_p25', 'low_p50', 'low_p75', 'low_mean', 'low_std',
                        'jd_count_1', 'mid_p25', 'mid_p50', 'mid_p75', 'mid_mean', 'mid_std',
                        'jd_count_2', 'high_p25', 'high_p50', 'high_p75', 'high_mean', 'high_std']
```

```python
city_features['year'] = 2021
city_features['half_year'] = 0
city_features_needed = city_features[cities_old.columns]
city_features_needed.fillna(0, inplace=True)
```

```python
city_features_needed.to_csv('../city_features.csv', index=False)
```
